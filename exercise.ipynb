```python
# input[1,3,2,4,1] output[1,3,3,-1,-1] find the index for the next greater number for each element in the array, if not exist, return -1
def suffix(nums): 
    array = [-1] * len(nums)
    # find the next greater element's index for each element in nums
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[j] > nums[i]:
                array[i] = j
                break
    return array 

suffix([1,3,2,4,1])
```
